## Density Estimation
We wish to build a density esstimation of a dataset $X$ with $N$ observations and for each observation a new $x$ determine the probability of it originating from the same distribution as the dataset.

### Problem
The dataset and the classes therein may follow a weirdly shaped distribution, and the solution is to construct a Gaussian Mixture Model (GMM) from multiple Multivariate Gaussians (MVG). An MVG is parametrized as follows: $N(x|\mu, \Sigma)$, \mu being the distributions mean, and \Sigma being the covariance matrix which determines the shape of the distribution.

### GMM
A GMM is defined as $p(x|\mu, \Sigma) = \sum_{k=1}^{K} \pi_k \cdot N(x | \mu_k, \Sigma_k)$
Where $\pi_k$ is the weighting of the MVG $k$.

### Latent variables
* Latent variables (unobserved) $Z \in R^{N \times K}$ where $K$ is the number of mixture components. 
* The entry $z_ik$ defines $x_i$'s membership of the mixture component $k$. If $z_{ik} = 1$ then $x_i$ belongs to component $k$.
* $\sum z_i = 1$, meaning only one entry is 1 - the rest are zero.

### EM Algorithm
* Objective: Estimate parameters $\theta = (\pi, \mu, \Sigma)$ which maximize the log-likehood the dataset $X$, ie. $p(X|\theta)$
* It does so by iteratively estimating the parameters $\theta$ and updating them to maximize the log-likelihood.
* Log-Likelihood definition: $L(X | \theta) = log \:p(X | \theta) = \sum_{i=1}^{N} log \:p(x_i | \theta) = \sum_{i=1}^{N} log [\sum_{k=1}^{K} \pi_k \cdot N(x_i | \theta)]$

### Notation
* $\gamma_{ik}$: Probability for datapoint $x_i$ belonging to mixture component $k$
* $N_k$: Total mass of component $k$
* $N$: Total mass of all $k$ components 
* $u_k$: Mean of component $k$
* $\Sigma_k$: Covariance matrix of component $k$
* $\pi_k$: Weight of component $k$

### Definitions
* $\gamma_{ik}$ is computed using Bayes' theorem for $x_i$ belonging to component $k$, which is the case for $z_k = 1$.
    * $p(z_k=1 x_i) = \frac{ p(x_i | z_k = 1) \cdot p(z_k = 1) }{ \sum_{j=1}^{K} p(x_i | z_j=1) \cdot p(z_j=1) } $
    * $= \frac{ \pi_k \cdot N(x_i | \mu_k, \Sigma_k) }{ \sum_{j=1}^{K} \pi_j \cdot N(x_i | \mu_j, \Sigma_j }$
* $N_ k = \sum_{i=1}^{N} \gamma_{ik}$

Empirical measures:
* $\mu_k = \frac{1}{N_k} \sum_{i=1}^{N} \gamma_{ik} \cdot x_i$
* $\Sigma_k = \frac{1}{N_k} \sum_{i=1}^{N} \gamma_{ik} \cdot (x_i-\mu_k)(x_i-\mu_k)^T$
* $\pi_k = \frac{N_k}{N}$

### EM Algorithm Pseudocode
* Input: X, K
* Init parameters $\theta = (\mu_k, \Sigma_k, \pi_k)$ for $k=1...K$
* While L not converged, do:
    * $\gamma_{ik} := \frac{ \pi_k \cdot N(x_i | \mu_k, \Sigma_k) }{ \sum_{j=1}^{K} \pi_j \cdot N(x_i | \mu_j, \Sigma_j }$
    * $N_ k = \sum_{i=1}^{N} \gamma_{ik}$
    * $\mu_k := \frac{1}{N_k} \sum_{i=1}^{N} \gamma_{ik} \cdot x_i$
    * $\Sigma_k := \frac{1}{N_k} \sum_{i=1}^{N} \gamma_{ik} \cdot (x_i-\mu_k)(x_i-\mu_k)^T$
    * $\pi_k := \frac{N_k}{N}$
    * $L = \sum_{i=1}^{N} log [\sum_{k=1}^{K} \pi_k \cdot N(x_i | \theta)]$
* return $\theta$

```python
import numpy as np
from scipy.stats import multivariate_normal as mvn

def EM(X, K, eps=0.01, max_iter=100):
    N, M = X.shape
    pi = np.ones(K) / K # Start with even weighting for each component
    mu = np.random.random((K, M))
    sigma = np.array([np.eye(M)] * K)
    
    # Previous log likelihood value
    ll_old = -np.inf
    
    for i in range(max_iter):
        #### E-Step ####
        gamma = np.zeros((K, N))
        for k in range(K):
            gamma[k] = pi[k] * mvn(mean=mu[k], cov=sigma[k]).pdf(X)
        gamma /= gamma.sum(0)
        
        #### M-Step ####
        # Update mass vector
        mass = gamma.sum(axis=1) 
        
        # Update means
        # Add new axis to 'mass' in order to perform vector division
        mu = np.dot(gamma, X) / mass[:, None] 
        
        # Update weights
        pi = mass / N
        
        # Update covariance matrices
        sigma = np.zeros((K, M, M))
        for k in range(K):
            for i in range(N):
                ys = np.reshape(X[i] - mu[k], (2,1))
                sigma[k] += gamma[k, i] * np.dot(ys, ys.T)
            sigma[k] /= mass[k]
                
        # Compute log-likelihood
        ll_new = 0
        for p, m, s in zip(pi, mu, sigma):
            ll_new += p * mvn(m, s).pdf(X)
        ll_new = np.log(ll_new).sum()

        print("log-likelihood:", ll_new)
        # Check if converged
        if abs(ll_new - ll_old) < eps: 
            break
        ll_old = ll_new
    
    return ll_new, pi, mu, sigma
```

## 2 Mixture components

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(123)

# create data set
n = 1000
K = 2
_mus = np.array([[0,4], [-2,0]])
_sigmas = np.array([[[3, 0], [0, 0.5]], [[1,0],[0,2]]])
_pis = np.array([0.6, 0.4])

xs = np.concatenate([np.random.multivariate_normal(mu, sigma, int(pi*n))
                    for pi, mu, sigma in zip(_pis, _mus, _sigmas)])
ax = plt.subplot(111)
plt.scatter(xs[:,0], xs[:,1], alpha=0.2)
ax.axes.set_aspect('equal')
plt.show()
```

```python
# Two mixture components (K = 2)
K = 2
ll, pi, mu, sigma = EM(xs, K)
```
