### Problem 2: Apply K-Means to the Image, each pixel being a 3d vector

```python
def k_means(X, K=2, max_iter=100):    
    # Init means
    U = X[np.random.choice(np.arange(X.shape[0]), K, replace=False)]
    Z_old = np.zeros(X.shape[0])
    
    for i in range(max_iter):
        # Assign data points to nearest centroid using euclidean dist
        distances = np.sum((X[:, np.newaxis] - U[np.newaxis, :])**2, axis=2) # Squared
        Z = np.argmin(distances, axis=1)
        
        # Compute new cluster means. 
        # (Since clusters dont have the same length, loop cant be vectorized)
        valid_clusters = np.zeros(K, dtype=int) # Keep track of zero clusters
        for k in range(K):
            cluster_points = X[Z == k]
            if len(cluster_points) > 0:
                U[k] = cluster_points.mean(axis=0)
                valid_clusters[k] = 1
        
        # Remove empty clusters
        U = U[valid_clusters.nonzero()]
        
        # If nothing changed, convergence has been reached
        if np.all(Z == Z_old): 
            break
            
        # Overwrite old means
        Z_old = Z
    
    print("Finished in %i iterations." % i)
    print("Dropped %i clusters." % (K - valid_clusters.sum()))
    return U, Z
```

```python
def entropy(z):
    probs = np.array([(z == i).mean() for i in set(z)])
    return -np.sum(probs * np.log2(probs))

def compress_img(im, K):
    X = im.reshape(-1, 3).astype('float32')/255
    u, z  = k_means(X, K)
    im_reconstruction = u[z].reshape(im.shape)
    
    plt.figure(figsize=(8, 8))
    plt.imshow(im_reconstruction)
    plt.show()
    
    print("Entropy per pixel:", entropy(z))
```

```python
compress_img(im, 4)
```
