## 1 Probability Refresher
A couple has two children, each of them being independently a boy or a girl with 50% probability. Compute the probabilities of the following events.

1. At least one of the children is a girl.
2. Both children are girls.
3. Both children are girls given that the first born is a girl.
4. Both children are girls given that one of them is a girl.
5. Both children are girls given that one of them is a girl named Cassiopeia.
Note: Cassiopeia is an extremely rare name with a frequency of less than 1 in 1,000,000.

````
X = {x_1, x_2}: seq of children

N = {n_1, n_2}: seq of names

c: cassiopeia = 1e-6

g: child is girl = 0.5

b: child is boy = 0.5
```
1: At least one girl

$p(x_1=g \cup x_2=g) = p(x_1=g) + p(x_2=g) -  p(x_1=g) \cap p(x_2=g) = 1 - p(x_1 = b, x_2 = b)$

$= 1 - 0.5 \cdot 0.5 = 0.5 + 0.5 - 0.5^2 = 0.75$

2: Both are girls
$p(x_1 = g \cap x_2=g) = p(x_1 = g) \cdot p(x_2 = g) = 0.5^2 = 0.25$

3: Both girls, given first is girl
$p(x_1=g, x_2=g|x_1=g) = \frac{p(x_1=g, x_2=g) \cap p(x_1=g)}{p(x_1=g)}$

We use that $p(x_1=g, x_2=g) \cap p(x_1=g) = p(x_1=g, x_2=g) = p(x_1=g) p(x_2=g)$, thereby:

$p(x_1=g, x_2=g|x_1=g) = \frac{p(x_1=g) p(x_2=g)}{p(x_1=g)} = p(x_2=g) = 0.5$

4: Both are girls, given either is girl
* $p({x_1=g \cap x_2=g} | {x_1=g \cup x_2=g}) = \frac{p(x_1=g \cap x_2=g)}{p(x_1=g \cup x_2=g)} = \frac{1/4}{3/4} = 1/3$

5: Both are girls given one is named cassiopeia (Only girls can be named Cassiopeia)

Both are girls: $p(x_1=g \cap x_2=g)$

Problem 2: Bayes Rule

* $p(d) = 0.01$: Prob of random person having the disease
* $p(\bar{d}) = 1 - p(d) = 0.99$: Probability of no disease
* $p(t|d) = 0.99$: Positive test score, given disease
* $p(\bar{t}|\bar{d}) = 0.99$: Negative test score, given no disease
* $p(t|\bar{d}) = 1 - p(\bar{t}|\bar{d}) = 0.01$: Positive test score, given no disease



Goal: $p(d|t) = \frac{p(t|d)p(d)}{p(t)}$

$p(t) = p(t|d)p(d) + p(t|\bar{d}) p(\bar{d}) = 0.99 \cdot 0.01 + 0.99 \cdot 0.01$

$p(d|t) = \frac{0.99 \cdot 0.01}{0.99 \cdot 0.01 + 0.99 \cdot 0.01} = 1/2$

Problem 2: Apply K-Means to the Image, each pixel being a 3d vector

`````python
def k_means(X_, K=2, max_iter=100):
    X = X_.copy()
    means = np.zeros(K)
    prev_means = np.ones(K)
    
    # Randomize the data
    X.shuffle()
    
    # Split data into k clusters
    clusters = np.array_split(X, K)
    
    for i in range(max_iter):
        # Compute new cluster means. 
        # (Since clusters dont have the same length, loop cant be vectorized)
        means = np.array([np.mean(clusters[k]) for k in range(K)])
        
        # If nothing changed, convergence has been reached
        if np.all(means == prev_means): break
        
        # Reset clusters
        clusters = [None] * K
        
        # Assign data points to nearest centroid using euclidean dist
        for x in X:
            euc_norm = (x - means)**2
            cluster_idx = euc_norm.argmin()
            new_clusters[cluster_idx].append(x)

        # Overwrite old means
        prev_means = means
        
    return new_clusters
`````
