```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
# pip3 install Pillow
import os,sys
from PIL import Image
```

```python
# Load data

image_dir = "CroppedYale/"

files = os.listdir(image_dir)
n = len(files)
print("Loading " + str(n) + " images")

imgs = [Image.open(image_dir + files[i]) for i in range(n)]
```

```python
plt.imshow(imgs[0]) # Display first image
```

**1. Build a matrix collecting all images as its columns**

```python
# Assume all images have the same size
img0 = imgs[0]
width, height = img0.size
wh = width*height

# Compute input matrix X
X_list = [np.reshape(imgs[i].getdata(), (wh)) for i in range(n)]
X = np.array(X_list)
print(X.shape)
```

**2. Normalize all images by subtracting the mean**

```python
mX = X.mean(axis=0)
Xm = X - mX
Xm.shape
```

```python
img_s = Xm[0].reshape((height, width))
plt.subplot(1,2,1)
plt.imshow(imgs[0]) # Display first image

plt.subplot(1,2,2)
plt.imshow(img_s) # Display first image
```

**3. Perform PCA on the covariance matrix**

```python
Sigma = np.cov(Xm.T)
U,S,VT = np.linalg.svd(Sigma, full_matrices=False)
```

```python
#U,S,VT = np.linalg.svd(Xm, full_matrices=False)
```

**4. Visualize the 5 first principal components. What is the interpretation of these images?**

```python
rho = S**2 / np.sum(S**2)
r = np.arange(1, rho.size + 1)

# Plot variance explained
plt.figure(figsize=[10,5])
plt.plot(r,rho,"o-", markeredgecolor="black")
plt.title("Variance explained") 
plt.xlabel("Principal component")
plt.ylabel("Fraction␣of␣variance␣explained")
plt.xlim([0,10])
plt.show()
```

```python
cumulative = np.cumsum(rho)
plt.plot(r, cumulative , "o-", markeredgecolor="black") 
plt.title("Cumulative␣variance␣explained␣by␣principal␣components") 
plt.xlabel("Principal␣Direction(s)") 
plt.ylabel("Cumulative fraction of variance explained")
plt.ylim([0,1])
plt.xlim([0,10])
plt.show()
```

```python
rho[0:5].sum()
```

Observations:
 * The first principal component alone accounts for more than 80% of the variance.
 * In order to account for $\geq95\%$ of the variance, the images must be represented with at least 5 dimensions, which is quite lower than the original 4080 dimensions!

```python
img_u = U.T[0].reshape((height, width))

plt.figure(figsize=[10,10])
plt.subplot(1,3,1)
plt.imshow(imgs[0]) # Display first image

plt.subplot(1,3,2)
plt.imshow(img_s) # Display first image

plt.subplot(1,3,3)
plt.imshow(img_u) # Display first image
```

Compressions $Z = U\bar{X}^T$

```python
Uk = U.T[:5]
Zk = np.dot(Uk, Xm.T)
Zk.shape
```

Decompression: $X = U^TZ + \mu_X = U^T U\bar{X}^T + \mu_X = X$

```python
Xd = np.dot(Uk.T, Zk) + mX.reshape((height*width, 1))
```

```python
img_d = Xd.T[0].reshape((height, width))

plt.figure(figsize=[10,10])
plt.subplot(1,4,1)
plt.imshow(imgs[0]) # Display first image

plt.subplot(1,4,2)
plt.imshow(img_s) # Display first image

plt.subplot(1,4,3)
plt.imshow(img_u) # Display first image

plt.subplot(1,4,4)
plt.imshow(img_d) # Display first image
```
